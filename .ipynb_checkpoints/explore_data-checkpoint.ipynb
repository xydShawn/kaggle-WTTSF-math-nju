```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import operator
import re

%matplotlib inline
```

```python
df1=pd.read_csv("../data/train_1.csv")
```

```python
df1.info()

df1.Page
```

```python
nameArray = df1.Page.apply(str).values
wordCount = {}
for x in nameArray:
    for y in x.split("_"):
        if y in wordCount:
            wordCount[y] += 1
        else:
            wordCount[y] = 1
sorted_x = sorted(wordCount.items(), key=operator.itemgetter(1), reverse=True)                
```

```python
sorted_x
```

```python
wordCount2 = {}
for x in nameArray:
    for y in re.split('_|wikipedia|org', x):
        if y in wordCount2:
            wordCount2[y] += 1
        else:
            wordCount2[y] = 1
sorted_wc2 = sorted(wordCount2.items(), key=operator.itemgetter(1), reverse=True)       
```
